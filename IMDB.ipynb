```python
data=data.drop_duplicates()
data
```

```python
data.describe()
```

```python
plt.figure(figsize=(20,10),dpi=50)
data.groupby('Genre')['No_of_Votes'].mean().sort_values(ascending=False).head(15).plot(kind='area',color='purple')
plt.xlabel('Genre',size=10)
plt.ylabel('Votes',size=10)
plt.title('Top genres by votes',size=15)
plt.show()
```

```python
data['Gross'] = data['Gross'].fillna(0)

data.pivot_table(values=['Gross','No_of_Votes','Meta_score'], index='Released_Year', aggfunc='mean').sort_values(by=['Meta_score','No_of_Votes'],ascending=False).plot()
```

```python
data.groupby('Star1')['Meta_score'].mean().sort_values(ascending=True)
```

```python
data['Runtime']=data['Runtime'].str.replace('min','')

data['Runtime']=pd.to_numeric(data['Runtime'])

top_20_movies = data.sort_values(by='Runtime',ascending=False)[['Series_Title','Runtime']].head(20)
```

```python
sns.barplot(x='Series_Title',y='Runtime',data =top_20_movies)
plt.title('Top 20 length movies')
plt.xticks(rotation=90)
```
